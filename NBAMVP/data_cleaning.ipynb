```python
players["Player"] = players["Player"].str.replace("*", "", regex=False) 
```

```python
players.head(50)
```

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Team"] == "TOT"]
        row["Team"] = df.iloc[-1,:]["Team"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_row)
```
